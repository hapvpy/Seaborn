# 1. Coordinate properties


## 1-1. x, y, xmin, xmax, ymin, ymax
좌표 속성은 플롯에서 마크가 그려지는 위치를 결정합니다. 공식적으로 `x` 좌표는 수평 위치이고 `y` 좌표는 수직 위치입니다.
일부 마크는 하나 또는 두 변수 모두에 대해 범위(즉, `min`, `max`) 매개변수화를 허용합니다. 다른 사람들은 `x`와 `y`를 받아들일 수 있지만 `baseline` 매개변수를 사용하여 범위를 표시할 수도 있습니다. 레이어의 `orient` 매개변수에 따라 이것이 작동하는 방식이 결정됩니다.

변수에 숫자 데이터가 포함되어 있지 않으면 화면에 데이터를 그릴 수 있도록 해당 스케일이 변환을 적용합니다. 예를 들어, `Nominal` 척도는 각각의 고유한 범주에 정수 인덱스를 할당하고 `Temporal` 척도는 날짜를 참조 "epoch"으로부터의 일 수로 나타냅니다.

`Continuous` 척도는 데이터 값과 공간 위치 간에 비선형 변환을 적용할 수도 있습니다.

# 2. Color properties

## 2-2. color, fillcolor, edgecolor

모든 마크에는 `color`가 지정될 수 있으며 대부분 마크의 "edge" 색상과 "fill" 색상을 구별합니다. 종종 단순히 `color`를 사용하면 둘 다 설정되지만 보다 구체적인 속성을 사용하면 추가 제어가 가능합니다.

image.png

색상 속성이 매핑되면 기본 색상표는 배율 유형에 따라 다릅니다. 공칭 척도는 불연속적이고 정렬되지 않은 색조를 사용하는 반면 연속 척도(시간적 척도 포함)는 순차적인 그라데이션을 사용합니다.
image.png

- 색상 스케일은 `'viridis'`, `'rocket'` 또는 `'deep'`과 같은 팔레트 이름으로 매개변수화됩니다. 일부 팔레트 이름에는 단순 그라데이션(예: `'dark:blue'`) 또는 cubehelix 시스템(예: `'ch:start=.2,rot=-.4'`)을 비롯한 매개변수가 포함될 수 있습니다. 적절한 선택에 대한 지침은 색상 팔레트 자습서를 참조하십시오.
- 연속 스케일은 스케일이 보간해야 하는 색상 튜플로 매개변수화할 수도 있습니다. 명목 척도를 사용하는 경우 팔레트 이름(필요한 경우 개별적으로 샘플링됨), 개별 색상 값 목록 또는 데이터 값을 색상에 직접 매핑하는 사전을 제공할 수 있습니다.
- 개별 색상은 다양한 형식으로 지정될 수 있습니다. 여기에는 현재 색상 주기(`'C0'`), 단일 문자 속기(`'b'`), 회색조 값(`'.4'`), RGB 16진수 코드(`'#4c72b0'`), X11 색상 이름(`'seagreen'`)에 대한 색인 참조가 포함됩니다. ') 및 XKCD 색상 조사 이름(`'purpleish'`):

## alpha, fillalpha, edgealpha
`alpha` 속성은 마크의 불투명도를 결정합니다. 오버플로팅의 경우 알파를 낮추면 밀도를 나타내는 데 도움이 될 수 있습니다.
image.png

  알파 속성 매핑은 마크가 겹치지 않는 경우에도 중요성을 전달하고 두 가지 변수를 나타내는 `color` 스케일과 결합할 수 있기 때문에 유용할 수 있습니다. 또한 알파가 낮은 색상은 채도가 낮아져 더 큰 채워진 표시(예: 막대)의 모양을 개선할 수 있습니다.

  `color`와 마찬가지로 일부 마크는 추가 제어를 위해 별도의 `edgealpha` 및 `fillalpha` 속성을 정의합니다.

# Style properties

## 3-1. fill

`fill` 속성은 가장자리와 내부가 구분되는 마크와 관련이 있으며 내부가 보이는지 여부를 결정합니다. 부울 상태입니다. `fill`은 `True` 또는 `False`로만 설정할 수 있습니다.
image.png

## 3-2. marker

  `marker` 속성은 점 표시 및 일부 선 표시와 관련이 있습니다. 마커를 지정하기 위한 API는 matplotlib API 문서 `matplotlib.markers`에 자세히 설명된 것처럼 매우 유연합니다.

  여러 간단한 문자열 코드를 사용하여 마커를 지정할 수 있습니다.

## 3-3. linestyle, edgestyle
linestyle 속성은 라인 마크와 관련이 있으며 edgestyle 속성은 "가장자리"가 있는 여러 마크와 관련이 있습니다. 두 속성 모두 온-오프 세그먼트 측면에서 선의 "대시"를 결정합니다.

대시는 소수의 속기 코드('-', '--', '-.' 및 ':')를 사용하거나 프로그래밍 방식으로 튜플(on, off, ...)을 사용하여 지정할 수 있습니다. 튜플 사양에서 단위는 선폭과 같습니다.

# Size properties

## 4-1. pointsize

`pointsize` 속성은 개별 데이터 포인트에서 마커를 표시할 수 있는 점 표시 및 선 표시와 관련이 있습니다. 단위는 마크의 직경(포인트)에 해당합니다.

`pointsize`는 기본적으로 데이터의 제곱근으로 크기가 조정되므로 크기는 면적이 아닌 직경으로 표시됩니다.

image.png

## 4-2. linewidth

`linewidth` 속성은 라인 마크와 관련이 있으며 두께를 결정합니다. 값은 음수가 아니어야 하며 포인트 단위가 있어야 합니다.

image.png

## 4-3. edgewidth


`edgewidth` 속성은 `linewidth`와 유사하지만 선이 아닌 가장자리/채우기가 있는 마크에 적용됩니다. 또한 척도에서 사용될 때 다른 기본 범위를 갖습니다. 단위는 동일합니다.

image.png

## 4-4. stroke

`stroke` 속성은 `edgewidth`와 비슷하지만 점 표시가 채우기가 아닌 획으로 정의될 때 적용됩니다. 또한 기본 축척 범위가 약간 다르지만 그 외에는 유사하게 작동합니다.

image.png

# Text properties

## 5-1. align, valign


`halign` 및 `valign` 속성은 텍스트 마크의 수평 및 수직 정렬을 제어합니다. 수평 정렬 옵션은 `'left'`, `'right'`, `'center'`이고, 수직 정렬 옵션은 `'top'`, `'bottom'`, `'center'`, `'baseline'`, `'center_baseline'`입니다.

image.png

## 5-2. fontsize


`fontsize` 속성은 텍스트 표시의 크기를 제어합니다. 값에는 다음과 같은 포인트 단위가 있습니다.

image.png

## 5-3. offset

`offset` 속성은 텍스트 마크와 앵커 위치 사이의 간격을 제어합니다. `center` 정렬을 사용하지 않을 때(즉, 왼쪽/오른쪽 또는 위/아래를 사용할 때) 적용됩니다. 값에 포인트 단위가 있습니다.

image.png

# Other properties

## 6-1. text


`text` 속성은 텍스트 마크의 내용을 설정하는 데 사용됩니다. 항상 문자 그대로 사용되며(매핑되지 않음) 필요한 경우 문자열로 캐스트됩니다.

## 6-2. group

`group` 속성은 마크의 모양에 대해 아무 것도 변경하지 않지만 변환이 독립적으로 작동해야 하는 추가 데이터 하위 집합을 정의한다는 점에서 특별합니다.
